NumPy is a Python library for working with arrays and matrices of numerical data. It provides powerful tools for manipulating arrays, performing mathematical operations on arrays, and working with complex data structures.

NumPy arrays are homogeneous, meaning all elements in an array must have the same data type. This allows for efficient computation, especially for large arrays.

NumPy also includes tools for linear algebra, Fourier analysis, and random number generation, among other things. It is a fundamental library for scientific computing and is widely used in the data science and machine learning communities.

```python
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt

# Load image from URL into a NumPy array
url = 'https://upload.wikimedia.org/wikipedia/commons/f/fa/Grayscale_8bits_palette_sample_image.png'
with urllib.request.urlopen(url) as url_file:
    image = Image.open(url_file)
    img_arr = np.array(image)
print(img_arr.shape)
# Perform SVD on the image array
U, S, Vt = np.linalg.svd(img_arr, full_matrices=False)

# Use the top 10 singular values to reconstruct the image
k = 10
S_k = np.diag(S[:k])
reconstructed = np.dot(U[:, :k], np.dot(S_k, Vt[:k, :]))

# Display the original image and the reconstructed image with 10 singular values
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[1].imshow(reconstructed)
axes[1].set_title('Reconstructed Image with 10 Singular Values')
plt.show()
```

```python
import numpy as np

# define two arrays
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 6, 8, 10])

# compute the cosine similarity
cosine_similarity = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
print("Cosine similarity:", cosine_similarity)

# compute the correlation coefficient
correlation_coefficient = np.corrcoef(a, b)[0, 1]
print("Correlation coefficient:", correlation_coefficient)

# compute the variance
variance = np.var(a)
print("Variance:", variance)

# compute the 25th and 75th quantiles of a
quantile_25 = np.quantile(a, 0.25)
quantile_75 = np.quantile(a, 0.75)
print("25th quantile:", quantile_25)
print("75th quantile:", quantile_75)
```

```python
import numpy as np

# Create two arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Compute Euclidean distance
euclidean_dist = np.sqrt(np.sum((a - b)**2))
print("Euclidean distance:", euclidean_dist)

# Compute L1 distance
l1_dist = np.sum(np.abs(a - b))
print("L1 distance:", l1_dist)

# Compute L2 distance
l2_dist = np.sqrt(np.sum((a - b)**2))
print("L2 distance:", l2_dist)

# Compute L-infinity distance
linf_dist = np.max(np.abs(a - b))
print("L-infinity distance:", linf_dist)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values 
x = np.linspace(0, 10, 50)

# Generate the y values
y1 = x + np.random.normal(0, 0.5, size=50)
y2 = x + np.random.normal(0, 3, size=50)
y3 = x + np.random.normal(0, 15, size=50)
y4 = -x + np.random.normal(0, 2, size=50)

# Compute the correlation coefficient for each pair of points
corr_coef1 = np.corrcoef(x, y1)[0, 1]
corr_coef2 = np.corrcoef(x, y2)[0, 1]
corr_coef3 = np.corrcoef(x, y3)[0, 1]
corr_coef4 = np.corrcoef(x, y4)[0, 1]

# Fit a linear regression line to each pair of points
slope1, intercept1 = np.polyfit(x, y1, 1)
slope2, intercept2 = np.polyfit(x, y2, 1)
slope3, intercept3 = np.polyfit(x, y3, 1)
slope4, intercept4 = np.polyfit(x, y4, 1)

# Create a 2x2 plot to display the points and their linear regression line
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].scatter(x, y1)
axs[0, 0].plot(x, slope1*x + intercept1, color='red')
axs[0, 0].set_title(f'Correlation coefficient: {corr_coef1:.2f}')

axs[0, 1].scatter(x, y2)
axs[0, 1].plot(x, slope2*x + intercept2, color='red')
axs[0, 1].set_title(f'Correlation coefficient: {corr_coef2:.2f}')

axs[1, 0].scatter(x, y3)
axs[1, 0].plot(x, slope3*x + intercept3, color='red')
axs[1, 0].set_title(f'Correlation coefficient: {corr_coef3:.2f}')

axs[1, 1].scatter(x, y4)
axs[1, 1].plot(x, slope4*x + intercept4, color='red')
axs[1, 1].set_title(f'Correlation coefficient: {corr_coef4:.2f}')

plt.show()
```

```python
import pandas as pd
import numpy as np

# create a dataframe with daily data
dates = pd.date_range(start='2021-06-01', end='2022-12-31', freq='D')
data = pd.DataFrame({'date': dates, 'value': np.random.randint(1, 100, len(dates))})

# add month and year columns to dataframe
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year

print(data)
```

```python
# aggregate data by month and year
agg_data = data.groupby(['year', 'month'])['value'].sum().reset_index()
print(agg_data)
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import cosine_similarity

# Load the iris dataset
iris = load_iris()
X = iris.data.copy()

# Define the sample and feature of interest
i = 10
j = 2

# Compute cosine similarity of ith sample with all samples (ignoring jth feature)
similarity = cosine_similarity(X[:, np.arange(X.shape[1]) != j], X[i, np.arange(X.shape[1]) != j].reshape(1, -1)).ravel()

# Find k similar samples
k = 5
k_similar_samples = np.argsort(similarity)[-k-1:-1]

# Fill jth feature with the average of jth feature of k similar samples
X[i, j] = np.mean(X[k_similar_samples, j])

print(f"Original value of X[{i}, {j}]: {iris.data[i, j]}")
print(f"New value of X[{i}, {j}]: {X[i, j]}")
```

```python
import numpy as np
from scipy.stats import entropy

# Create a set of discrete data between 0 and 4
data = np.random.randint(0, 3, size=1000)
#data = np.random.randint(0, 4, size=1000)

# Compute the probability distribution of the data
p_data = np.zeros(4)
for i in range(4):
    p_data[i] = np.mean(data == i)

# Compute the entropy of the data
entropy_data = entropy(p_data, base=4)

print("Probability distribution of the data: ", p_data)
print("Entropy of the data: ", entropy_data)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.datasets import load_iris

# load iris dataset
iris = load_iris()
X = iris.data

# normalization using StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# normalization using MinMaxScaler
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)

# normalization using RobustScaler
robust_scaler = RobustScaler()
X_robust = robust_scaler.fit_transform(X)

# print mean and standard deviation of normalized features
print("StandardScaler mean:", np.mean(X_std, axis=0))
print("StandardScaler std:", np.std(X_std, axis=0))
print("MinMaxScaler mean:", np.mean(X_minmax, axis=0))
print("MinMaxScaler std:", np.std(X_minmax, axis=0))
print("RobustScaler mean:", np.mean(X_robust, axis=0))
print("RobustScaler std:", np.std(X_robust, axis=0))
```
